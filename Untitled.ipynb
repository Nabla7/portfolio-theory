# Portfolio Optimization Using Lagrange Multipliers

## Introduction

In modern portfolio theory, the primary goal is to construct a portfolio that maximizes return while minimizing risk. The two key metrics we focus on are:

- **Expected Return ($\mu_p$)**: The weighted average of the expected returns of the assets in the portfolio.
- **Risk ($\sigma_p$)**: The standard deviation of the portfolio's returns, representing the portfolio's volatility.

We will use matrix algebra and the Lagrange multiplier method to determine the optimal weights that minimize the portfolio's risk for a given level of expected return.

## Expected Return of the Portfolio

The expected return of the portfolio, $\mu_p$, is defined as:

$$
\mu_p = \sum_{i=1}^{N} w_i \mu_i
$$

where:
- $w_i$ is the weight of asset $i$ in the portfolio,
- $\mu_i$ is the expected return of asset $i$,
- $N$ is the total number of assets in the portfolio.

## Variance and Standard Deviation of the Portfolio

The variance of the portfolio, $\sigma_p^2$, measures the risk associated with the portfolio and is given by:

$$
\sigma_p^2 = \mathbf{w}^\top \Sigma \mathbf{w}
$$

where:
- $\mathbf{w}$ is the vector of portfolio weights,
- $\Sigma$ is the covariance matrix of asset returns.

The standard deviation of the portfolio, which represents the risk, is:

$$
\sigma_p = \sqrt{\sigma_p^2}
$$

## Covariance Matrix

The covariance matrix $\Sigma$ contains variances along the diagonal and covariances off-diagonal. For a portfolio of three assets, the covariance matrix is:

$$
\Sigma = \begin{pmatrix}
\sigma_1^2 & \sigma_{12} & \sigma_{13} \\
\sigma_{12} & \sigma_2^2 & \sigma_{23} \\
\sigma_{13} & \sigma_{23} & \sigma_3^2
\end{pmatrix}
$$

where:
- $\sigma_i^2$ is the variance of asset $i$,
- $\sigma_{ij}$ is the covariance between assets $i$ and $j$.

## Optimization Using Lagrange Multipliers

To find the optimal portfolio weights that minimize variance subject to the constraint that all weights sum to 1, we use the Lagrange multiplier method. The Lagrangian function is defined as:

$$
\mathcal{L}(\mathbf{w}, \lambda) = \mathbf{w}^\top \Sigma \mathbf{w} + \lambda \left( \sum_{i=1}^{N} w_i - 1 \right)
$$

where:
- $\lambda$ is the Lagrange multiplier.

### Step 1: Setting Up the Lagrangian

The Lagrangian combines the objective function (the portfolio variance) with the constraint (the sum of the weights equaling 1). Our goal is to minimize the Lagrangian.

### Step 2: Deriving the First-Order Conditions

To find the minimum, we take the derivative of the Lagrangian with respect to each weight $w_i$ and the Lagrange multiplier $\lambda$, and set them to zero:

$$
\frac{\partial \mathcal{L}}{\partial w_i} = 2 \sum_{j=1}^{N} \Sigma_{ij} w_j + \lambda = 0 \quad \text{for all } i
$$

$$
\frac{\partial \mathcal{L}}{\partial \lambda} = \sum_{i=1}^{N} w_i - 1 = 0
$$

### Step 3: Solving the System of Equations

This gives us a system of $N+1$ equations (one for each weight and one for the constraint). We solve this system to find the optimal weights $\mathbf{w}^*$ and the Lagrange multiplier $\lambda$.

## Efficient Frontier

The efficient frontier represents the set of optimal portfolios that offer the highest expected return for a given level of risk. By solving the optimization problem for different levels of expected return, we can plot the risk-return combinations to visualize the trade-offs.

## Minimum Variance Portfolio

The minimum variance portfolio is the portfolio on the efficient frontier with the lowest possible risk. This portfolio is particularly important for risk-averse investors.

## Conclusion

In this analysis, we applied matrix algebra and the Lagrange multiplier method to optimize the portfolio, minimizing risk for a given expected return. The efficient frontier was plotted to provide a visual representation of the trade-off between risk and return, guiding investors to make informed decisions based on their risk tolerance.

```python
from tqdm import tqdm
import sympy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('bank_data.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Calculate returns
returns = np.log(data / data.shift(1)).dropna()

# Calculate annualized mean returns and covariance matrix
mean_returns = returns.mean() * 252  # Assuming 252 trading days in a year
cov_matrix = returns.cov() * 252

# Number of assets
n = len(mean_returns)

# Define the symbols for the weights and the Lagrange multipliers
weights = sp.symbols(f'w1:{n+1}')
lambda_ = sp.symbols('lambda')
gamma = sp.symbols('gamma')

# Portfolio variance function
portfolio_variance = sum(weights[i] * weights[j] * cov_matrix.iloc[i, j] 
                         for i in range(n) for j in range(n))

# Portfolio expected return
portfolio_return = sum(weights[i] * mean_returns[i] for i in range(n))

# Target return (vary this to get the efficient frontier)
target_returns = np.linspace(mean_returns.min(), mean_returns.max(), 100)

# Store risk and return values for plotting
risks = []
returns_ = []

# Use tqdm to display progress
for target in tqdm(target_returns, desc="Calculating Efficient Frontier"):
    # Lagrangian function with the return constraint
    Lagrangian = portfolio_variance + lambda_ * (sum(weights) - 1) + gamma * (portfolio_return - target)

    # Take the partial derivatives of the Lagrangian
    partial_derivatives = [sp.diff(Lagrangian, w) for w in weights]
    partial_derivatives.append(sp.diff(Lagrangian, lambda_))
    partial_derivatives.append(sp.diff(Lagrangian, gamma))

    # Solve the system of equations
    solution = sp.solve(partial_derivatives, weights + (lambda_, gamma))

    # Extract the weights and calculate risk and return
    opt_weights = [solution[w] for w in weights]
    opt_return = sum(opt_weights[i] * mean_returns[i] for i in range(n))
    opt_variance = sum(opt_weights[i] * opt_weights[j] * cov_matrix.iloc[i, j] for i in range(n) for j in range(n))
    opt_risk = sp.sqrt(opt_variance)
    
    risks.append(float(opt_risk.evalf()))
    returns_.append(float(opt_return.evalf()))

# Plot the efficient frontier
plt.figure(figsize=(10, 7))
plt.plot(risks, returns_, 'b-', label='Efficient Frontier')
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return')
plt.title('Efficient Frontier Using Lagrange Multipliers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Load the data from the uploaded CSV file
data = pd.read_csv('bank_data.csv')

# Convert 'Date' column to datetime and set it as index
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Calculate returns
returns = np.log(data / data.shift(1)).dropna()

# Calculate annualized mean returns and covariance matrix
mean_returns = returns.mean() * 252  # Assuming 252 trading days in a year
cov_matrix = returns.cov() * 252

# Function to calculate portfolio performance
def portfolio_annualized_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(weights * mean_returns)
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))
    return std, returns

# Function to minimize portfolio variance
def minimize_variance(weights, mean_returns, cov_matrix):
    return portfolio_annualized_performance(weights, mean_returns, cov_matrix)[0] ** 2

# Optimize portfolio for minimum variance
def optimize_portfolio(mean_returns, cov_matrix):
    num_assets = len(mean_returns)
    args = (mean_returns, cov_matrix)
    constraints = ({'type': 'eq', 'fun': lambda x: np.sum(x) - 1})
    bounds = tuple((0, 1) for asset in range(num_assets))
    
    result = minimize(minimize_variance, num_assets*[1./num_assets,], args=args, 
                      method='SLSQP', bounds=bounds, constraints=constraints)
    
    return result

# Run optimization
opt_results = optimize_portfolio(mean_returns, cov_matrix)
opt_weights = opt_results.x
opt_std, opt_return = portfolio_annualized_performance(opt_weights, mean_returns, cov_matrix)

print("Optimal Weights:", opt_weights)
print("Portfolio Standard Deviation (Risk):", opt_std)
print("Portfolio Expected Return:", opt_return)

# Function to display the efficient frontier
def display_efficient_frontier(mean_returns, cov_matrix):
    results = np.zeros((3, 10000))
    weights_record = []
    
    for i in range(10000):
        weights = np.random.random(len(mean_returns))
        weights /= np.sum(weights)
        weights_record.append(weights)
        
        portfolio_std_dev, portfolio_return = portfolio_annualized_performance(weights, mean_returns, cov_matrix)
        
        results[0,i] = portfolio_std_dev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - 0) / portfolio_std_dev  # Sharpe ratio with 0 risk-free rate
    
    plt.figure(figsize=(10, 7))
    plt.scatter(results[0,:], results[1,:], c=results[2,:], cmap='YlGnBu', marker='o')
    plt.title('Efficient Frontier')
    plt.xlabel('Risk (Standard Deviation)')
    plt.ylabel('Return')
    plt.colorbar(label='Sharpe ratio')
    
    min_variance_port = optimize_portfolio(mean_returns, cov_matrix)
    min_var_std, min_var_return = portfolio_annualized_performance(min_variance_port.x, mean_returns, cov_matrix)
    plt.scatter(min_var_std, min_var_return, marker='*', color='r', s=200, label='Minimum Variance Portfolio')
    
    plt.legend(labelspacing=0.8)
    plt.show()

# Display the efficient frontier
display_efficient_frontier(mean_returns, cov_matrix)
```
